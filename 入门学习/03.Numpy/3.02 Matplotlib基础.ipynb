## plot 二维图

* plot(y)
* plot(x, y)
* plot(x, y, format_string)



```python
%matplotlib inline
x = linspace(0, 2 *pi, 50)
#只给定 y 值，默认以下标为 x 轴：
plot(sin(x))
```

```python
# 给定 x 和 y 值：
plot(x, sin(x))
```

```python
# 多条线：
plot(x, sin(x), x, sin(2 * x))
```

```python
# 使用字符串，给定线条参数：
plot(x, sin(x), 'r-^')
```

```python
# 多条线
plot(x, sin(x), 'b-o',
    x, sin(2 * x), 'r-')
```

## Scatter散点图

* scatter(x, y)
* scatter(x, y, size)
* scatter(x, y, size, color)

假设我们想画二维散点图：

```python
plot(x, sin(x), 'o')
```

可以使用 scatter 达到同样的效果：

```python
scatter(x, sin(x))
```

scatter函数与Matlab的用法相同，还可以指定它的大小，颜色等参数

```python
x = rand(200)
y = rand(200)
size = rand(200) * 30
color = rand(200)
scatter(x, y, size, color)

#显示颜色条
colorbar()
```

## 多图

使用**figure()**命令产生新的图像：

```python
t = linspace(0, 2*pi, 50)
x = sin(t)
y = cos(t)
plot(x)
#产生新的图像
figure()
plot(y)
```

或者使用 subplot 在一幅图中画多幅子图：

    subplot(row, column, index)

```python
subplot(1, 2, 1)
plot(x)
subplot(1, 2, 2)
plot(y)
```

## 向图中添加数据


默认多次 plot 会叠加：

```python
plot(x)
plot(y)
```

可以跟**Matlab**类似用 **hold(False)**关掉，这样新图会将原图覆盖：(3.X已经移除了hold)

```python
plot(x)
hold(False)
plot(y)
```

## 标签

可以在 plot 中加入 **label** ，使用 **legend** 加上图例：

```python
plot(x, label='sin')
plot(y, label='cos')
legend()
```

或者直接在 legend中加入：

```python
plot(x)
plot(y)
legend(['sin', 'cos'])
```

## 坐标轴，标题，网格

```python
%pylab
%matplotlib inline
```

```python
x = linspace(0, 2 *pi, 50)
plot(x, sin(x))
xlabel('radians')
# 设置字体大小
ylabel('amplitue', fontsize='large')
title('Sin(x)')
```

用**grid()** 来显示网格：

```python
plot(x, sin(x))
xlabel('radians')
ylabel('amplitude', fontsize='large')
title('Sin(x)')
grid()
```
